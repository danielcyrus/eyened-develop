```python
from sqlalchemy import select, func, update
from eyened_orm_internal.ergo import (
    Database,
    ImageInstance,
    Modality,
    Series,
    Study,
    Patient,
    Project
)
from eyened_orm.inference.inference import run_inference_for_images
```

```python
database = Database('/home/jose/vessels/eyened_platform/dev/prod.env')
session = database.create_session()
```

```python
# Sample 4 random images from the development database
stmt = (
    select(ImageInstance)
    .where(ImageInstance.Modality == Modality.ColorFundus)
    .order_by(func.random())
    .limit(4)
)
images = list(session.scalars(stmt).all())
print(f"Selected {len(images)} images")
```

```python
# Run inference with no cfi_cache_path
run_inference_for_images(session, images, device=None)
```

```python
# Print the image IDs
image_ids = [im.ImageInstanceID for im in images]
print("Image IDs:")
for img_id in image_ids:
    print(img_id)
```
