```python
from eyened_orm import FormSchema, Database, EntityType

db = Database('/path/to/your/.env')
```

```python
import json
schema = json.load(open("schema.json"))
```

## Adding to the database

```python

with db.get_session() as session:
    formschema = FormSchema(
        SchemaName="referable_glaucoma_grading_justification",
        Schema=schema,
        EntityType=EntityType.StudyEye,
    )
    session.add(formschema)
    session.commit()
```

## To update the schema
Though we recommend making a new version in case of changes

```python


schema = json.load(open("schema.json"))
with db.get_session() as session:
    formschema = FormSchema.by_id(session, 10)
    formschema.Schema = schema
    session.commit()
```
